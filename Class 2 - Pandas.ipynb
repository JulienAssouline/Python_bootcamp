```python
countries = pd.read_csv("country names - Sheet1.csv")
```

```python
countries["id_3char"] = countries["id_3char"].str.upper()
```

```python
countries.head()
```

```python
# rename these columns!!!! 
```

```python
countries.columns = ["id", "code", "Name"]
```

```python
countries.head()
```

```python
df_merged = df.merge(countries, left_on = "country destination", right_on = "code")
```

```python
df_merged.head()
```

```python
df_merged.drop(["id", "code"], 1, inplace = True)
```

```python
df_merged.head()
```

```python
df_China = df_merged[df_merged["Name"] == "China"]
```

```python
china_2015_exp_val = df_China[df_China["year"] == 2015]["export_val"]
```

```python
df_merged.dtypes
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-darkgrid')
my_dpi=96
plt.figure(figsize=(480/my_dpi, 480/my_dpi), dpi=my_dpi)


for key, group in df_merged.groupby("Name"):
    plt.plot(group.year, group.export_val, marker='', label = key, color="lightgrey")
# plt.set_xlabel('Year')
# plt.set_ylabel('export_val')
plt.xlabel("Year")
plt.ylabel("export_val")
plt.ticklabel_format(style='plain')

plt.plot(df_China.year, df_China.export_val, color="red")
plt.text(2015.2, china_2015_exp_val, 'China', horizontalalignment='left', size='medium', color='red')
plt.title("China is trading more with North Korea than anyone else", loc='left', fontsize=12, fontweight=0, color='black')
plt.show()
```
